Find the largest of three numbers using conditions.

```python
def largest(a,b,c):
  if a>b and a>c:
    print(a,"is the largest")
  elif b>a and b>c:
    print(b,"is the largest")
  else:
    print(c,"is the largest")
a=int(input("Enter a number: "))
b=int(input("Enter a number: "))
c=int(input("Enter a number: "))
largest(a,b,c)
```

#Task 3


Write a program to calculate the factorial of a number using a loop.

```python
num=int(input("Enter a number: "))
fact=1
i=1

while i<=num:
    fact *=i
    i +=1

print("Factorial:",fact)
```

Print table of the given number

```python
num=int(input("Enter a number: "))

for i in range(1, 11):
    print(f"{num} x{i}={num * i}")
```

Count Digits in Number and their Sum

```python

num=input("Enter a number: ")
digits_count=len(num)
digits_sum=sum(int(d) for d in num)
print("Digits count:", digits_count)
print("Sum of digits:", digits_sum)
```
