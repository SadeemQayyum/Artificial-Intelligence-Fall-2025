```python
import numpy as np
# Array of 150 random integers between 1 and 1000
arr=np.random.randint(1,1000,size=150)
even_num=arr[arr%2==0]
print("Even Numbers\n",even_num)
# to print prime numbers only
def is_prime(num):
    if num<2:
        return False
    for i in range(2,int(num**0.5)+1):
        if num%i==0:
            return False
    return True

prime_num=np.array([num for num in arr if is_prime(num)])
print("Prime Numbers\n",prime_num)
def get_multiples(array,number):
    return array[array %number==0]
n=int(input("\nEnter a number to find its multiples: "))
multiples=get_multiples(arr, n)
print(f"\nMultiples of {n}:\n",multiples)


```
