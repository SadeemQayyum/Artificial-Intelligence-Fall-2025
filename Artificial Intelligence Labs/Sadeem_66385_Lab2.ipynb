1. append() - Adds an element at the end of the list

```python
fruit_list=["Apple","Banana","Mango","Grapes"]
fruit_list.append("Orange")
print(fruit_list)
```

2. clear() - Removes all the elements from the list

```python
fruit_list.clear()
print(fruit_list)
```

3. copy() - Returns a copy of the list

```python
fruit_list=["Apple","Banana","Mango","Grapes"]
fruit_list2=fruit_list.copy()
print(fruit_list)
```

4. count() - Returns the number of elements with the specified value

```python
fruit_list=["Apple","Banana","Mango","Grapes","Apple"]
fruit_list.count("Apple")
```

5. extend() - Add the elements of a list (or any iterable), to the end of the current list

```python
fruit_list=["Apple","Banana","Mango","Grapes"]
fruit_list2=["Orange","Kiwi"]
fruit_list.extend(fruit_list2)
print(fruit_list)
```

6. index() - Returns the index of the first element with the specified value

```python
fruit_list=["Apple","Banana","Mango","Grapes"]
fruit_list.index("Mango",2)
```

7. insert() - Adds an element at the specified position

```python
fruit_list=["Apple","Banana","Mango","Grapes"]
fruit_list.insert(2,"Orange")
print(fruit_list)
```

8. pop() - Removes the element at the specified position

```python
fruit_list=["Apple","Banana","Mango","Grapes"]
fruit_list.pop(2)
print(fruit_list)
```

9. remove() - Removes the item with the specified value

```python
fruit_list.remove("Apple")
print(fruit_list)
```

11. sort() - Sorts the list

```python
fruit_list=["Apple","Banana","Mango","Grapes"]
fruit_list.sort()
number_list=[9,3,7,1,3]  #ascending order
number_list.sort()
print(fruit_list)
print(number_list)
```

10. reverse() - Reverses the order of the list

```python
fruit_list.sort(reverse=True)
number_list.sort(reverse=True) #descending order
print(fruit_list)
print(number_list)
```
