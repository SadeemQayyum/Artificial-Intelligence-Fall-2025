```python
# Function to evaluate the Tic Tac Toe grid
def evaluate(grid):
    # Checking Rows for victory
    for row in range(3):
        if grid[row][0]==grid[row][1]==grid[row][2]:
            if grid[row][0]=='x':
                return 10    # X (maximizer) wins
            elif grid[row][0]=='o':
                return -10   # O (minimizer) wins

    # Checking Columns for victory
    for col in range(3):
        if grid[0][col]==grid[1][col] == grid[2][col]:
            if grid[0][col]=='x':
                return 10
            elif grid[0][col]=='o':
                return -10

    # Checking Diagonals for victory
    if grid[0][0]==grid[1][1]==grid[2][2]:
        if grid[0][0]=='x':
            return 10
        elif grid[0][0]=='o':
            return -10

    if grid[0][2]==grid[1][1]==grid[2][0]:
        if grid[0][2]=='x':
            return 10
        elif grid[0][2]=='o':
            return -10

    # If no winner
    return 0


# Example Tic Tac Toe Grids (Different Scenarios)
grid1=[
    ['x', 'x', 'x'],
    ['o', '_', 'o'],
    ['_', '_', '_']
]  # X wins (first row)

grid2=[
    ['o', 'x', 'x'],
    ['x', 'o', '_'],
    ['_', '_', 'o']
]  # O wins (main diagonal)

grid3=[
    ['x', 'o', 'x'],
    ['o', 'x', 'o'],
    ['o', 'x', 'o']
]  # Draw


# Evaluate and print results for all 3 grids
for i, grid in enumerate([grid1, grid2, grid3], start=1):
    print(f"\n===== Grid {i} =====")
    for row in grid:
        print(row)
    score = evaluate(grid)

    # Print results
    if score == 10:
        print(" X wins! Evaluation Value:", score)
    elif score == -10:
        print(" O wins! Evaluation Value:", score)
    else:
        print(" Draw or Game in Progress. Evaluation Value:", score)
```
